```python
import numpy as np
import matplotlib.pyplot as plt

# Crie um vetor de tempo de -2 a 2 segundos com intervalos de 0,01 segundos
t = np.arange(-2, 2, 0.01)

# Crie o sinal x(-0.5t)
sinal = np.sin(0.5 * np.pi * t)

# Plote o sinal
plt.figure(figsize=(8, 4))
plt.plot(t, sinal)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Sinal x(-0.5t)')
plt.grid(True)
plt.show()
```
